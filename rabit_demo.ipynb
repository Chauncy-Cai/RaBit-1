```python
import numpy as np
import meshio
import openmesh as om
import trimesh
from rabit_np import RaBitModel
```

```python
np.random.seed(2)

rabit = RaBitModel()

# pose_shape: [23, 3]
# beta_shape: [500]
# theta = np.random.rand(*(23, 3))*0.1
theta = np.zeros((23, 3))
beta = np.random.rand(*(500,)) * 10 - 5
beta[10:] = 0
trans = np.zeros(rabit.trans_shape)

rabit.set_params(beta=beta, pose=theta, trans=trans)

mesh = trimesh.Trimesh(vertices=rabit.verts, faces=rabit.quads.reshape(-1,4))
mesh.show()
```
